```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python
Lista = [] # filtro y generar muertes totales a la semana
with open('Def.csv', newline='') as File:
    datos2 = csv.reader(File, delimiter=';')
    
    for row in datos2:
        Lista.append(row[1])
        Lista.append(row[12])
```

```python
Cantidad = []
for f in Fechas:
    datos = Lista.count(f)
    Cantidad.append(datos)
```

```python
fallecidos_semana = 0
n = 0
Fallecidos = []

for c in Cantidad:
    n = n + 1
    fallecidos_semana = fallecidos_semana + c
    
    if n/7 == 1:
        Fallecidos.append(fallecidos_semana)
        n = 0
        fallecidos_semana = 0
```

```python
fig, ax = plt.subplots()  

ax.plot(Date,Fallecidos, linestyle = "solid")
fig.autofmt_xdate()
plt.xticks(np.arange(0, len(x)+50, 5))
plt.show()
```
